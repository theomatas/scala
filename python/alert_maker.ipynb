```python
from pandas import read_csv, DataFrame
from random import random
dt = read_csv('light_2.csv')
dt.sample(3)
```

```python
dt_clean = DataFrame()
save = ['Summons Number','Plate ID','Registration State','Issue Date', 'Violation Description','Vehicle Make', 'Vehicle Expiration Date', 'Vehicle Color']
for name in save:
    dt_clean[name] = dt[name]
dt_clean.sample(3)
```

```python
dt_clean['alert'] = [ 1 if random() < 0.015 else 0 for idx in range(len(dt_clean))]
```

```python
dt_clean.sample(3)
```

```python
dt_clean['alert'].value_counts()
```

```python
spliter = lambda cell: cell.split('-',1) if type(cell) == str else ""
temp = DataFrame(tuple(map(spliter,dt_clean['Violation Description'])))
dt_clean[['Violation code','Violation Description']] = temp
dt_clean.sample(5)
```
